```python
fig, ax = plt.subplots(figsize=(10, 6))

branche = ['Landbrug/skov/fiskeri', 'Industri', 'Forsyningsvirksomhed', 'Øvrige virksomheder']
udledning = [12681, 5579, 7565, 8831]

ax.bar(branche, udledning);
```
